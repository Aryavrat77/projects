```python
# takes the percentage change of "df" Dataframe and assigns it to a new Dateframe named "df_rtn".

df_rtn = df.pct_change()
```

https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.shift.html

```python
# shift it by one day, so that today's value shows YESTERDAY's return,


df_rtn_prev = df_rtn.shift(1)
```

```python
df_rtn_prev.head(7)  # The first two rows should have NaN, and Jan 3 should have Jan 2's return.
```

https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.rolling.html

```python
# gets a rolling sum of prior (including the current date) 5 days. 

df_5d_rolling = df_rtn_prev.rolling(5).sum()
```

```python
df_5d_rolling.head(7) # The first six rows should have Nan, and Jan 9 should be the sum of the entries
# Jan 3, 4, 7, 8, 9 of df_rtn_prev, which is actually the sum of the entries Jan 2, 3, 4, 7, 8 of df_rtn.
```

```python
df_5d_rolling_trimmed = df_5d_rolling.dropna()  # It is convenient to drop the rows with missing data.
df_5d_rolling_trimmed.head()
```

```python

df_5d_rolling.dropna(inplace=True)
df_5d_rolling.head()
```

```python
mr_sig_raw = - df_5d_rolling
# The idea is to use the return over the prior 5 days as a trading signal.
# Our hypothesis is that we should have short-term mean reversion ("mr") of stock returns;
# that is why we apply a negative sign to our signal.  We short stocks that (relatively speaking)
# have gone up a lot in the previous 5 days, and we buy the stocks that have gone down.
```

https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.mean.html

```python
# assigns the cross sectional mean return to a pandas time series called "mr_sig_mean".

mr_sig_mean = mr_sig_raw.mean(axis=1)
```

```python
mr_sig_mean.tail(3)
```

https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.std.html

```python
# assigns the cross sectional standard deviation to a pandas time series called "mr_sig_std".

mr_sig_std = mr_sig_raw.std(axis=1)
```

```python
mr_sig_std.tail(3)
```

https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.sub.html    
https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.div.html

```python
# normalize our signals to have mean 0, variance 1.
# mr_sig_norm = (mr_sig_raw - mr_sig_mean) / mr_sig_std 

mr_sig_norm = mr_sig_raw.sub(mr_sig_mean, axis=0).div(mr_sig_std, axis=0)
```

```python
mr_sig_norm.tail(3)  # should get panel of standard deviations.
```

```python
# We want to do something similar to get a momentum signal.  We want to consider the last year of
# returns (we'll use 252 trading days), but excluding the most recent month (21 prior trading days + today).
mom_sig_raw = df_rtn.shift(22).rolling(231).sum().dropna()
mom_sig_raw.head(3)
```

```python
# normalize the momentum signal to have mean 0 and variance 1,

mom_sig_norm = mom_sig_raw.sub(mom_sig_raw.mean(axis=1), axis=0).div(mom_sig_raw.std(axis=1), axis=0)
```

```python
mom_sig_norm.tail(3)  # should get panel of standard deviations.
```

```python
mom_rtn = (mom_sig_norm/(574-1) * df_rtn).dropna().sum(axis=1)
# What we have done used mom_sig_norm / (N-1), with N=574 to use as a position size in dollars.
```

```python
# Execute this cell.
mom_rtn_mean = mom_rtn.mean() * 252     # We express our mean return on an annual basis.
mom_rtn_std = mom_rtn.std() * 252**0.5  # We express our standard deviation on an annual basis.
mom_SR = mom_rtn_mean / mom_rtn_std     # This is the annual Sharpe ratio of our strategy.
print(mom_rtn_mean, mom_rtn_std, mom_SR)  # should get a Sharpe ratio of +0.83
```

```python
mr_rtn = (mr_sig_norm/(574-1) * df_rtn).truncate(before='2014-01-02').sum(axis=1)
# We truncate the time before Jan 2014 so that both the momentum series and the mean
# reversion series have the same date range (2 years of data).
```

```python
# Sharpe ratio of the mean reversion strategy. Get a Sharpe ratio of +0.87
mr_rtn_mean = mr_rtn.mean() * 252     # We express our mean return on an annual basis.
mr_rtn_std = mr_rtn.std() * 252**0.5  # We express our standard deviation on an annual basis.
mr_SR = mr_rtn_mean / mr_rtn_std     # This is the annual Sharpe ratio of our strategy.
print(mr_rtn_mean, mr_rtn_std, mr_SR)
```

```python
both_rtn = 0.5*mom_rtn + 0.5*mr_rtn  # This is the return of a 50/50 portfolio of both strategies.
```

```python
both_rtn_mean = both_rtn.mean() * 252     # We express our mean return on an annual basis.
both_rtn_std = both_rtn.std() * 252**0.5  # We express our standard deviation on an annual basis.
both_SR = both_rtn_mean / both_rtn_std     # This is the annual Sharpe ratio of our strategy.
print(both_rtn_mean, both_rtn_std, both_SR) # Higher Sharpe ratio
```

```python
# Execute this cell.
plt.figure()
mom_rtn.cumsum().plot(label="Momentum")
mr_rtn.cumsum().plot(label="Mean Reversion")
both_rtn.cumsum().plot(label="50/50 Both")
plt.legend()
plt.show()
```

```python

plt.figure()
(mom_rtn.cumsum()/mom_rtn_std*0.15).plot(label="Momentum (std=0.15)")
(mr_rtn.cumsum()/mr_rtn_std*0.15).plot(label="Mean Reversion (std=0.15)")
(both_rtn.cumsum()/both_rtn_std*0.15).plot(label="50/50 Both (std=0.15)")
plt.legend()
plt.show()
```
